# Reverse String
Given a string, reverse the string.

Example: input_string = "abcd". Output_string = "dcba"

**Note:** strings are immutable unlike lists. So you cannot modify the string in-plance. So you may want to create an empty string and add the reversed characters to it one by one.

```
def reverse_string(input_str):
  new_str = ""
  for i in range(len(input_str)):
    new_str += input_str[len(input_str)-1-i]
  return new_str
```

```
input_string = "abcd"
print("given string: "+input_string)
reversed_string = reverse_string(input_string)
print("reversed string: "+reversed_string)
```

# Check Palindrome
Check if a given string is a palindrome using recursion.

```
def is_palindrome(string):
  if len(string) == 1:
    return True
  return (string[0] == string[-1]) and is_palindrome(string[1:-1])
```

```
input_string = "madam"
print(input_string + " is a palindrome: ", is_palindrome(input_string))
input_string = "anirban"
print(input_string + " is a palindrome: ", is_palindrome(input_string))
```

# All permutations of a given string
Given a string, return all possible permutations of the given string.

Example: input = "abc". output = ["abc", "acb", "cab", "cba", "bca", "bac"]

```
def find_permutations(input_str, index):
  if index >= len(input_str):
    return [""]
  small_output = find_permutations(input_str, index+1)
  print(small_output)

  current_char = input_str[index]
  output_list = list()
  for permutation in small_output:
    for ind in range(len(small_output[0])+1):
      new_permutation = permutation[0:ind] + current_char + permutation[ind:]
      output_list.append(new_permutation)
  return output_list
```

```
input_string = "abc"
permuted_strings = find_permutations(input_string, 0)
print(permuted_strings)
```
