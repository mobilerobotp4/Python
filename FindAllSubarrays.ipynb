# Find All Subarrays

```
def find_all_subarrays(array):
  def comp_subarray(array, index):
    if index >= len(array):
      return [[]]

    small_output = comp_subarray(array, index+1)
    output = list()
    
    for element in small_output:
      output.append(element)
    
    for element in small_output:
      current = list()
      current.append(array[index])
      current.extend(element)
      output.append(current)
    
    return output

  output_array = comp_subarray(array, 0)
  return output_array
```

```
input_array = [9, 12, 15]
output_arr = find_all_subarrays(input_array)
print(output_arr)
```
