# Count Primes 
Count the number of prime numbers less than a non-negative number, `n`.

**Example:**
```
Input: 10
Output: 4
Explanation: There are 4 prime numbers less than 10, they are 2, 3, 5, 7.
```

````
def count_primes(num):
  prime_set = set()
  prime_set.add(2)
  if num == 2:
    return prime_set

  for ele in range(3, num):
    is_prime = True
    for i in prime_set:
      if (ele%i == 0):
        is_prime = False
        break
    if is_prime:
      prime_set.add(ele)

  return prime_set
````

````
num = 10
prime_set = count_primes(num)
print("All the prime numbers below {} is: {}".format(num, prime_set))
````
