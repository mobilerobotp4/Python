# Letter Combinations Phone Number

Given a string containing digits from `2-9` inclusive, return all possible letter combinations that the number could represent.

The mapping between the digits and the letters are as following,
`mapping = [".", ".", "abc", "def", "ghi", "jkl", "mno", "pqrs", "tuv", "wxyz"]`.

Note that, in the above mapping, digits `0` and `1` haas no valid mapping.

**Example:**
```
Input: "23"
Output: ["ad", "ae", "af", "bd", "be", "bf", "cd", "ce", "cf"].
```
Note:
Although the above answer is in lexicographical order, your answer could be in any order you want.


````
def letter_combinations_for_digits(digit_str):
  combs = list()
  mapping = {"0":".", "1":".", "2":"abc", "3":"def", "4":"ghi", "5":"jkl", "6":"mno", "7":"pqrs", "8":"tuv", "9":"wxyz"}

  def get_combination(d_str, index, mapping):
    if index == len(d_str) - 1:
      small_output = [i for i in mapping[d_str[index]]]
      return small_output

    small_output = get_combination(d_str, index+1, mapping)
    output = list()
    
    current = mapping[d_str[index]]
    for c in current:
      for id, ele in enumerate(small_output):
        output.append(c+ele)

    return output

  combs = get_combination(digit_str, 0, mapping)
  return combs
````

````
# Test case 1
digit_str = "23"
combinations = letter_combinations_for_digits(digit_str)
print(combinations)
````
